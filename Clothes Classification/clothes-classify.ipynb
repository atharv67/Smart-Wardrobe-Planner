```python
X=[]
y=[]
newsize=(128,128)
for i in range(len(tshirt_images)):
    img=cv2.imread(os.path.join(tshirt_dir,tshirt_images[i]))
#     plt.imshow(test_img)
    img=cv2.GaussianBlur(img,(3,3),9)
    X.append(cv2.resize(img,newsize,interpolation=cv2.INTER_AREA))
    y.append(0)
for i in range(len(shirt_images)):
    img=cv2.imread(os.path.join(shirt_dir,shirt_images[i]))
#     plt.imshow(test_img)
    img=cv2.GaussianBlur(img,(3,3),9)
    X.append(cv2.resize(img,newsize,cv2.INTER_AREA))
    y.append(1)
for i in range(len(shorts_images)):
    img=cv2.imread(os.path.join(shorts_dir,shorts_images[i]))
#     plt.imshow(test_img)
    img=cv2.GaussianBlur(img,(3,3),9)
    X.append(cv2.resize(img,newsize,cv2.INTER_AREA))
    y.append(2)
for i in range(len(pants_images)):
    img=cv2.imread(os.path.join(pants_dir,pants_images[i]))
#     plt.imshow(test_img)
    img=cv2.GaussianBlur(img,(3,3),9)
    X.append(cv2.resize(img,newsize,cv2.INTER_AREA))
    y.append(3)
for i in range(len(jackets_images)):
    img=cv2.imread(os.path.join(jackets_dir,jackets_images[i]))
#     plt.imshow(test_img)
    img=cv2.GaussianBlur(img,(3,3),9)
    X.append(cv2.resize(img,newsize,cv2.INTER_AREA))
    y.append(4)
```

```python
X,y=shuffle(X_array,y,random_state=1)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.03,random_state=1)
```

```python
# X_train=np.repeat(X_train[..., np.newaxis], 3, -1)
# X_test=np.repeat(X_test[..., np.newaxis], 3, -1)
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
def model(input_size):
  model=tf.keras.Sequential()
  mobile=tf.keras.applications.MobileNet(
    weights='imagenet',include_top=False,input_shape=input_size, 
    pooling='none'
)

  mobile.trainable=False
  model.add(mobile)
  model.add(tf.keras.layers.Flatten())
  model.add(tf.keras.layers.Dense(15,activation='relu'))
  model.add(tf.keras.layers.Dense(10,activation='relu'))
  model.add(tf.keras.layers.Dense(5,activation='softmax'))
  model.summary()
  return model
```

```python
checkpoint_path='/kaggle/working/model1'
checkpoint_dir = os.path.dirname(checkpoint_path)
```

```python
cp1_callback = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path,
                                                 save_weights_only=True,
                                                 verbose=1)
```

```python
optimizer = tf.keras.optimizers.Adam(lr=0.001)
```

```python
mod1=model((128,128,3))
mod1.compile(optimizer=optimizer,loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
mod1.fit(X_train,y_train,verbose=2,epochs=150,validation_split=0.05,batch_size=100,callbacks=[cp1_callback])
```

```python
mod1.evaluate(X_test,y_test)
```
